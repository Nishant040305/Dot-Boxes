### Time calculation wrapper

```python
def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        if len(args) > 0:
            N = args[0]
        else:
            N = kwargs.get("N", None)

        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()

        elapsed = end_time - start_time
        print(f"Time Executed: {elapsed:.4f} s.")
        if N is not None:
            normalized_time = elapsed / (2 * N * (N + 1))
            print(f"Time Per move: {normalized_time:.8f} s")
        return result
    return wrapper
```

```python
@timer
def init_game(N,depth=3):
    env = BaseBoardEnv(N)
    simulator = SimulateAlphaBetaVSAlphaBeta(env,depth1=depth,depth2=depth,parallel=True)
    result = simulator.simulate()
    print(f"Result: {result[0]} {result[1]}")
    return None
```
